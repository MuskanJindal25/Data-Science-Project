Out of 16,533 values, contact_adr2 and violation_sthigh have more than 70% null values. So we drop those as well

Imputing such large numbers of nan values in two features wouldn't be helpful. So we drop those as well. We can see that sam_id has no null values. This would be good news as this corresponds to ID in another dataset, which has details about addresses

Our current idea is to do analysis based on location, so null values in attributes longitude and latitude wouldn't be helpful to us. Our current visualization plans revolve around plotting the violation description on the city map using latitude, longitude, violation_zip and contact_addr1 values to see if we can find a trend between violation and the neighbourhood.

```python
data.drop(['violation_sthigh', 'contact_addr2', 'violation_state', 'contact_zip', 'contact_state', 'contact_city', 'violation_suffix', 'ticket_no'], axis = 1, inplace= True)

data.dropna(subset=['longitude', 'latitude', 'violation_zip', 'contact_addr1'], inplace=True)
```

```python
#description and code seem to have a one to many relationship
# Check for unique values in 'description' and 'code'
mapping = data.groupby('description')['code'].unique()

# Print the mapping
for key, values in mapping.items():
    print(f"{key} -> {list(values)}")
```

```python
mapping = data.groupby('code')['description'].unique()

# Print the mapping
for key, values in mapping.items():
    print(f"{key} -> {list(values)}")
```

We can club sub sections into a single section for ease of analysis. Like 9a, 9b, 9c to 9.
